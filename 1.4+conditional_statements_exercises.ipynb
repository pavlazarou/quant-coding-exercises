#### ðŸ“Š **Exercise 1: Trading Signal Generation**

Create a trading signal system based on price and volume conditions:
- **Current Price**: $175.50
- **Target Price**: $180.00
- **Volume**: 1,500,000 shares
- **Minimum Volume**: 1,000,000 shares

**Generate signals based on these conditions:**
- If price > target AND volume > minimum: "STRONG_BUY"
- If price > target OR volume > minimum: "BUY"
- If price < target AND volume < minimum: "SELL"
- Otherwise: "HOLD"

**Mathematical Logic:**
$$\text{Signal} = \begin{cases}
\text{STRONG\_BUY} & \text{if } P > T \text{ and } V > V_{min} \\
\text{BUY} & \text{if } P > T \text{ or } V > V_{min} \\
\text{SELL} & \text{if } P < T \text{ and } V < V_{min} \\
\text{HOLD} & \text{otherwise}
\end{cases}$$

**ðŸ’¡ Hint:** Use nested if/elif/else statements with AND/OR operators.


```python
# Exercise 1 Solution
current_price = 175.50
target_price = 180.00
volume = 1_500_000
min_volume = 1_000_000

if current_price > target_price and volume > min_volume:
    signal = "STRONG_BUY"
elif current_price > target_price or volume > min_volume:
    signal = "BUY"
elif current_price < target_price and volume < min_volume:
    signal = "SELL"
else:
    signal = "HOLD"

print(f"Trading Signal: {signal}")
```

#### ðŸ“ˆ **Exercise 2: Risk Management System**

Implement a risk management system with multiple conditions:
- **Portfolio Value**: $100,000
- **Position Size**: $15,000
- **Volatility**: 22%
- **Beta**: 1.3
- **Correlation**: 0.4

**Risk Rules:**
- If volatility > 25% OR beta > 1.5: "HIGH_RISK"
- If volatility > 20% AND beta > 1.2: "MEDIUM_RISK"
- If position > 20% of portfolio: "OVERSIZED"
- If correlation > 0.5: "CORRELATED"
- Otherwise: "ACCEPTABLE"

**Mathematical Logic:**
$$\text{Risk Level} = \begin{cases}
\text{HIGH\_RISK} & \text{if } \sigma > 0.25 \text{ or } \beta > 1.5 \\
\text{MEDIUM\_RISK} & \text{if } \sigma > 0.20 \text{ and } \beta > 1.2 \\
\text{OVERSIZED} & \text{if } \frac{P}{V} > 0.20 \\
\text{CORRELATED} & \text{if } \rho > 0.5 \\
\text{ACCEPTABLE} & \text{otherwise}
\end{cases}$$

**ðŸ’¡ Hint:** Use multiple if/elif statements with comparison and logical operators.


```python
# Exercise 2 Solution
portfolio_value = 100_000
position_size = 15_000
volatility = 0.22  # 22%
beta = 1.3
correlation = 0.4

# Calculate position size as percentage of portfolio
position_percentage = position_size / portfolio_value

if volatility > 0.25 or beta > 1.5:
    risk_level = "HIGH_RISK"
elif volatility > 0.20 and beta > 1.2:
    risk_level = "MEDIUM_RISK"
elif position_percentage > 0.20:
    risk_level = "OVERSIZED"
elif correlation > 0.5:
    risk_level = "CORRELATED"
else:
    risk_level = "ACCEPTABLE"

print(f"Risk Assessment: {risk_level}")
```

#### ðŸ”„ **Exercise 3: Portfolio Rebalancing Logic**

Create a portfolio rebalancing system with multiple conditions:
- **Target Allocation**: 60% stocks, 40% bonds
- **Current Allocation**: 65% stocks, 35% bonds
- **Rebalance Threshold**: 5%
- **Last Rebalance**: 30 days ago
- **Minimum Rebalance Period**: 30 days

**Rebalancing Rules:**
- If deviation > threshold AND time since last rebalance > minimum period: "REBALANCE"
- If deviation > threshold BUT time since last rebalance < minimum period: "WAIT"
- If deviation <= threshold: "NO_ACTION"
- If market is closed: "DEFER"

**Mathematical Logic:**
$$\text{Action} = \begin{cases}
\text{REBALANCE} & \text{if } |A_c - A_t| > T \text{ and } \Delta t > T_{min} \\
\text{WAIT} & \text{if } |A_c - A_t| > T \text{ and } \Delta t \leq T_{min} \\
\text{NO\_ACTION} & \text{if } |A_c - A_t| \leq T \\
\text{DEFER} & \text{if market closed}
\end{cases}$$

**ðŸ’¡ Hint:** Use nested conditions with AND/OR operators and comparison operators.


```python
# Exercise 3 Solution

# Given values
target_stocks = 0.60  # 60% target allocation for stocks
target_bonds = 0.40   # 40% target allocation for bonds
current_stocks = 0.65 # 65% current allocation in stocks
current_bonds = 0.35  # 35% current allocation in bonds
rebalance_threshold = 0.05  # 5% threshold
days_since_rebalance = 30
min_rebalance_period = 30
market_open = True

# Calculate deviation from target
stocks_deviation = abs(current_stocks - target_stocks)

# Check conditions and determine action
if not market_open:
    action = "DEFER"
elif stocks_deviation > rebalance_threshold:
    if days_since_rebalance > min_rebalance_period:
        action = "REBALANCE"
    else:
        action = "WAIT"
else:
    action = "NO_ACTION"

print(f"Portfolio Rebalancing Action: {action}")
```

#### ðŸŽ¯ **Exercise 4: Market Regime Detection**

Build a market regime detection system using multiple indicators:
- **Market Return**: 12% (annual)
- **Volatility**: 18%
- **VIX Level**: 22
- **Momentum**: 0.15
- **Sentiment**: 0.7

**Regime Classification:**
- **BULL_MARKET**: return > 10% AND volatility < 20% AND VIX < 25
- **BEAR_MARKET**: return < -5% OR volatility > 30% OR VIX > 35
- **VOLATILE_MARKET**: volatility > 25% OR VIX > 30
- **NEUTRAL_MARKET**: otherwise

**Mathematical Logic:**
$$\text{Regime} = \begin{cases}
\text{BULL\_MARKET} & \text{if } R > 0.10 \text{ and } \sigma < 0.20 \text{ and } \text{VIX} < 25 \\
\text{BEAR\_MARKET} & \text{if } R < -0.05 \text{ or } \sigma > 0.30 \text{ or } \text{VIX} > 35 \\
\text{VOLATILE\_MARKET} & \text{if } \sigma > 0.25 \text{ or } \text{VIX} > 30 \\
\text{NEUTRAL\_MARKET} & \text{otherwise}
\end{cases}$$

**ðŸ’¡ Hint:** Use if/elif/else with complex AND/OR conditions.


```python
# Exercise 4 Solution
# Define market indicators
market_return = 0.12  # 12% annual return
volatility = 0.18    # 18% volatility
vix_level = 22       # VIX at 22
momentum = 0.15      # Momentum indicator
sentiment = 0.7      # Market sentiment

# Determine market regime using conditions
if market_return > 0.10 and volatility < 0.20 and vix_level < 25:
    regime = "BULL_MARKET"
elif market_return < -0.05 or volatility > 0.30 or vix_level > 35:
    regime = "BEAR_MARKET"
elif volatility > 0.25 or vix_level > 30:
    regime = "VOLATILE_MARKET"
else:
    regime = "NEUTRAL_MARKET"

print(f"Current Market Regime: {regime}")
```
