#### ðŸ“Š **Exercise 1: Portfolio Analysis with Lists**

You have the following portfolio data:
- **Stocks**: `["AAPL", "GOOGL", "MSFT", "TSLA"]`
- **Shares**: `[150, 25, 200, 50]`
- **Prices**: `[175.50, 2850.75, 415.25, 890.30]`

**Calculate the total portfolio value and find the largest position.**

**Mathematical Formulas:**
$$\text{Total Value} = \sum_{i=0}^{n-1} \text{shares}[i] \times \text{prices}[i]$$
$$\text{Largest Position} = \max_{i} (\text{shares}[i] \times \text{prices}[i])$$

**ðŸ’¡ Hint:** Use list indexing and the `max()` function.


````python
# Exercise 1 Solution
# Your code here
Stocks = ["AAPL", "GOOGL", "MSFT", "TSLA"]
Shares = [150, 25, 200, 50]
Prices = [175.50, 2850.75, 415.25, 890.30]

total_value = (Shares[0]*Prices[0]) + (Shares[1]*Prices[1]) + (Shares[2]*Prices[2]) + (Shares[3]*Prices[3])

Largest_Position = max((Shares[0]*Prices[0]), (Shares[1]*Prices[1]), (Shares[2]*Prices[2]), (Shares[3]*Prices[3]))

print(f"Total Value: {total_value}")
print(f"Largest Position: {Largest_Position}")
````

#### ðŸ”— **Exercise 3: Portfolio Overlap with Sets**

You have two portfolios:
- **Portfolio A**: `{"AAPL", "GOOGL", "MSFT", "TSLA"}`
- **Portfolio B**: `{"MSFT", "TSLA", "NVDA", "META"}`

**Find the overlap and unique positions in each portfolio.**

**Set Operations:**
$$\text{Overlap} = A \cap B$$
$$\text{Unique to A} = A - B$$
$$\text{Unique to B} = B - A$$

**ðŸ’¡ Hint:** Use set intersection (`&`) and difference (`-`) operations.
 ex

````python
# Exercise 3 Solution
# Your code here
portfolio_a = {"AAPL", "GOOGL", "MSFT", "TSLA"}
portfolio_b = {"MSFT", "TSLA", "NVDA", "META"}

overlap = portfolio_a & portfolio_b

difference_a = portfolio_a - portfolio_b
difference_b = portfolio_b - portfolio_a

print(overlap)
print(difference_a)
print(difference_b)
````

#### ðŸ“ˆ **Exercise 6: Risk Analysis with Lists**

Calculate risk metrics from daily returns:
- **Returns**: `[0.025, -0.012, 0.008, 0.015, -0.003, 0.022, -0.005]`
- **Find**: Maximum gain, maximum loss, and average return

**Mathematical Formulas:**
$$\text{Max Gain} = \max(r_1, r_2, \ldots, r_n)$$
$$\text{Max Loss} = \min(r_1, r_2, \ldots, r_n)$$
$$\text{Average Return} = \frac{1}{n} \sum_{i=1}^{n} r_i$$

**ðŸ’¡ Hint:** Use `max()`, `min()`, and `sum()` functions with lists.


````python
# Exercise 6 Solution
# Your code here
returns = [0.025, -0.012, 0.008, 0.015, -0.003, 0.022, -0.005]

max_gain = max(returns)
min_gain = min(returns)
avg_return = (sum(returns))/len(returns)

print(max_gain)
print(min_gain)
print(avg_return)
````

#### ðŸŽ¯ **Exercise 8: Portfolio Rebalancing with Sets**

You have a target portfolio allocation:
- **Target Sectors**: `{"Technology": 40%, "Healthcare": 25%, "Finance": 20%, "Energy": 15%}`

Your current portfolio has these sectors:
- **Current Sectors**: `{"Technology", "Healthcare", "Technology", "Finance", "Technology"}`

**Find which sectors need rebalancing and which are missing.**

**Set Operations:**
$$\text{Missing Sectors} = \text{Target Sectors} - \text{Current Sectors}$$
$$\text{Overweight Sectors} = \text{Current Sectors} - \text{Target Sectors}$$

**Code Structure:**
```python
# Convert to sets for comparison
target_set = set(target_sectors.keys())
current_set = set(current_sectors)
missing = target_set - current_set
overweight = current_set - target_set
```

**ðŸ’¡ Hint:** Use set difference operations to find missing and excess sectors.


````python
# Exercise 8 Solution
# Your code here
target_sectors = {"Technology": 0.4, "Healthcare": 0.25, "Finance": 0.2, "Energy": 0.15}
current_sectors = {"Technology","Healthcare","Finance","Technology", "Technology"}

target_set = set(target_sectors.keys())
current_set = set(current_sectors)

missing = target_set - current_set
overweight = current_set - target_set

print(missing)
print(overweight)
````
